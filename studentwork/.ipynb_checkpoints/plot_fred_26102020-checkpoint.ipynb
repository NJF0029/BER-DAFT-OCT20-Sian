```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
import warnings
warnings.filterwarnings('ignore')
import matplotlib.pyplot as plt
import seaborn as sns 
%matplotlib inline
```

```python
data = pd.read_csv('mtcars.csv')
```

```python
data
```

```python
correlations = data.corr()
mask = np.zeros_like(correlations)
mask[np.triu_indices_from(mask)] = True
fig, ax = plt.subplots(figsize=(10, 8))
ax = sns.heatmap(correlations, mask=mask, annot=True)
plt.show()
```

```python
#strong negative correlation between horsepower and MPG
```

```python
sns.scatterplot(x='MPG', y='Horsepower', data=data)
```

```python
sns.scatterplot(x='MPG', y='Horsepower',hue='Cylinders', data=data)
```

```python
#sub hypothesis seems to be false as we can see that cars with less cylinders use less MPG
```
